```python
# import config # to hide TMDB API keys
import requests # to make TMDB API calls
import locale # to format currency as USD
import timeit
import json
import csv
locale.setlocale( locale.LC_ALL, '' )

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter # to format currency on charts axis
```

```python
diretorio_arquivo = r'~/Documents/Data Science/Projeto Integrador/Dados'
ArquivoJSON = '/bd_idIMDB_Rating_v6.csv'

#BaseJSON = open(ArquivoJSON)

bd_IdIMDB = pd.read_csv(diretorio_arquivo + ArquivoJSON, encoding='utf-8-sig') #,  lines=True, orient='records')
```

```python
b_filme = pd.DataFrame() # Inicia base para coletar resultados

for i, row in bd_IdIMDB.iterrows(): 
       
      print(i)
      print( 'id filme - ' + str(row['imdb_id']))
      
      b_detalhesfilmeID = pd.DataFrame() # Inicia base para coletar resultados
      
      key = 'k_78h194e0'  
      #key = 'k_fqw5ri9z'  
      #key = 'k_4qrglzzg'
      #key = 'k_123456'
      

      response = requests.get('https://imdb-api.com/en/API/Title/'+ key +'/' + str(row['imdb_id']) + 
                              '?/FullActor,Ratings,Wikipedia,' )
      
      #print ('https://api.themoviedb.org/3/movie/' + str(row['id']) + '?api_key=' + api_key+  '&language=en-US'  )
      
      #Convers√£o do arquivo JSON em um DataFrame
      js_filmes_detalhes = response.json()
      b_detalhesfilmeID = pd.json_normalize(js_filmes_detalhes)
      
      print(response.status_code)
      #print(str(l_filmes[i]))
      
      #Agrega os dados coletados na tabela existente
      
      b_filme = pd.concat([b_filme, b_detalhesfilmeID],join="outer") # Coleta resultado na base coletora
      
      

      # Atualiza o valor para o Campo Importado , definindo que o registro foi importado
      setattr(row,'importado',1)
      bd_IdIMDB.at[i,'importado']=1

      i += 1

      if (i == 5999): # or response.status_code != 200:
        #print(i)
        break
```

```python
nome_arquivo_descritivo = r'~/Documents/Data Science/Projeto Integrador/Dados/BaseCinemaIMDB' + dt_string + '.json' 

#nome_arquivo_descritivo = r'~/Documents/Data Science/DataWargling - Estudo/Arquivos processados/'+ nome_do_arquivo_entrada_saida_descritivo


b_filme.to_json( nome_arquivo_descritivo, orient='records', lines=True)
```
